# Collect images of Eastern Brown Snakes (Pseudonaja textilis) and Green Tree Snakes (Dendrelaphis punculata)

# Imports

```python
import os

from PIL import Image
from google_images_download import google_images_download
```

```python
response = google_images_download.googleimagesdownload()

def snakeDownload(species, venemous):
    arguments = {
        "keywords": species,
        "size": "large",
        "limit": 1300,
        "output": species,
        "print_urls": True,
        "chromedriver": "/bin/chromedriver",
        }

    paths = response.download(arguments)

    checkDownloads("downloads/" + species)

def checkDownloads(folder):
    for filename in os.listdir(folder):
        try:
            with Image.open(folder + "/" + filename) as im:
                print("")
        except:
            os.remove(folder + "/" + filename)
```

```python
snakeDownload(species="pseudonaja textilis", venemous=True)
snakeDownload(species="dendrelaphis punculata", venemous=False)
```
